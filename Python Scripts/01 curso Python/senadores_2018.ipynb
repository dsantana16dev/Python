# Análise dos dados de reembolso dos senadores 2018


```python
#importat bibliotecas
import pandas as pd
```

```python
#carregar o arquivo csv
df = pd.read_csv('despesa_ceaps_2018.csv', sep=';', encoding='latin1', skiprows=1, decimal=',')

df.head()
```

```python
# verificar tamanhos e tipos de dadps
df.info()
```

```python
#mostras uma analise quantitativa dos dados
df.describe()
```

```python
# total de reembolso

df['VALOR_REEMBOLSADO'].sum()
```

```python
# quantos reembolso por senador
df['SENADOR'].value_counts()
```

```python
# total de reembolso por senador
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum()
```

```python
# ordenar
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=True)
```

```python
# maiores reembolso
df.nlargest(10, 'VALOR_REEMBOLSADO')
```

```python
# 5 menores valores
df.nsmallest(5, 'VALOR_REEMBOLSADO')
```

```python
# quantidade de tipo de despesas
df['TIPO_DESPESA'].value_counts()
```
