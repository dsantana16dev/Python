```python
# importar bibliotecas
import pandas as pd
# importar matplotlib
%matplotlib inline
```

```python
# carregar o arquivo
df = pd.read_csv('cursos-prouni.csv', sep=';', encoding='cp1252')
df.head()
```

```python
# nudando coluna 'object' to 'float'
df['mensalidade'] = df['mensalidade'].str.replace(',','.')
df['mensalidade'] = df['mensalidade'].astype(float)
```

```python
df.dtypes
```

```python
# quais os 5 maiores mesalidades
df.nlargest(5, 'mensalidade')
```

```python
# quais são as 5 menores mensalidades
df.nsmallest(6, 'mensalidade')
```
