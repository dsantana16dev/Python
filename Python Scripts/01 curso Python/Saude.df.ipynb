```python
df1 = pd.read_csv('nascimento.csv', sep=';', encoding='cp1252')
df1.head(3)
df1.tail(3)
```

```python
# listar os hospitais presente nos dados
df1['estabelecimento_cnes'].unique()
```

```python
# plotar o gráfico por procedimento
df1['procedimento'].value_counts().plot.bar()
```

```python
# vamos criar um subconjunto dos dados originais
df_hospinf = df1[df1['estabelecimento_cnes'] == 'HOSPITAL MATERNO INFANTIL DE BRASILIA - HMIB']

# linas iniciais
df_hospinf.head()
```

```python
df_hospinf.tail(10)
```

```python
# ver linas aleatórias DF (pedaços)

df_hospinf.sample(5)
```

```python
# quantidade de procedimentos realizados

df_hospinf['procedimento'].value_counts()
```

```python
# podemos fatiar o DF com base em um pedaço de palavra
# str = string formula => df[df['coluna'].str.contains('xxx')].count()

df_hospinf[df_hospinf['procedimento'].str.contains('vagina')].count()
```

```python
# dividir o DF original apenas pelo procedimento de consulta

df_consulta = df1[df1['procedimento'] == 'Consulta Medica em Atenção Especializada']

df_consulta.head()
```
