# decisão - DESCARTAR - Relançamento  então transformar tudo em data

```python
# temos 8089 - criar novo DT sem os Relançamentos
df.shape
```

```python
df.head()
```

```python
dfn = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
dfn.dtypes
```

```python
#converter a coluna data de lançamento para datetime
dfn['Data de lançamento'] = pd.to_datetime(dfn['Data de lançamento'])
```

```python
#quais os anos tiveram mais filmes lançadosabs
dfn['Ano de exibição'].value_counts()
```

```python
dfn['Ano de exibição'].value_counts().plot.bar(title="lançamentos por ano")
```

```python
dfn['Renda (R$) no ano de exibição'].value_counts()
```

```python
# tirar os filmes que tem ND
dfn = dfn[dfn['Renda (R$) no ano de exibição'] != 'ND'].copy()
```

```python
# transformar OBJECT to Float
dfn['Renda (R$) no ano de exibição'] = dfn['Renda (R$) no ano de exibição'].astype(float)
```

```python
dfn.dtypes
```

```python
#qual o filme com maior renda
dfn[dfn['Renda (R$) no ano de exibição'] == dfn['Renda (R$) no ano de exibição'].max()]
```

```python
#qual o filme com 5 maiores renda
dfn.nlargest(3, 'Renda (R$) no ano de exibição' )
```

```python
dfn.nsmallest(15, 'Renda (R$) no ano de exibição')
```
