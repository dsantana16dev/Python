```python
import pandas as pd
import requests

url = 'https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA'

#requisitar pagina e quardar resposta
resposta = requests.get(url)

#exibir
#resposta.text (esse comando pega tudo que está na pagina)

#obter a resposta em formato.html
table = pd.read_html(resposta.text)

table
```

```python
# quero tabela com valores da final
#12 tabela
df= table[11]

df.head()
```
