Encontrar os numeros Nulos: não encontrei

```python
def show_nulls2016(rel2016):
    total_nan2016 = rel2016.isnull().sum()
    perc_nan2016 = total_nan2016/len(rel2016)*100
    return total_nan2016, perc_nan2016
```

```python
print('total_nulos 2016', show_nulls2016(rel2016=rel2016)[0])
```

```python
def show_nulls2015(rel2015):
    total_nan2015 = rel2015.isnull().sum()
    perc_nan2015 = total_nan2015/len(rel2015)*100
    return total_nan2015, perc_nan2015
```

```python
print('Total nulos 2015', show_nulls2015(rel2015=rel2015)[0])
```

```python
def show_nulls(df):
    total_nan = df.isnull().sum()
    perc_nan = total_nan/len(df)*100
    return total_nan, perc_nan
```

```python
print('total_nulos', show_nulls(df=df)[0])
```
