```python
def get_phrases():
    frases = []
    respuesta = input()
    while respuesta != 'basta':
        frases.append(respuesta)
        respuesta = input()
    return frases
```

```python
def filter_phrases(lista_frases):
    
    frases_con_a = []
    
    for item in lista_frases:
        if 'a' in item:
            frases_con_a.append(item)
        else:
            pass
    return frases_con_a
```

```python
def show_filtered_phrases(frases_filtradas):
    for frase in range(len(frases_filtradas)):
        print('Frase en posición {} --> {} '.format(frase, frases_filtradas[frase]))
```

```python
if __name__ == '__main__':
    todas_las_frases = get_phrases()
    frases_clean = filter_phrases(lista_frases=todas_las_frases)
    show_filtered_phrases(frases_filtradas=frases_clean)
```

Escreva uma função que receba uma sequência de números (pode ser uma lista ou uma tupla, e os números podem ser inteiros ou decimais) e retorna o número de números, o valor máximo, o valor mínimo e o valor médio.

```python
sec= (1, 89, 54, 98, 45, 23, 32.8, 1.3, 9.5)
print(sec)
```

```python
def sequence_status(input_sec):
    length = len(input_sec)
    max_value = max(input_sec)
    min_value = min(input_sec)
    suma = 0
    for item in input_sec:
        suma += item
    media = round(suma/length, 2)
    return length, min_value, max_value, media
```

```python
if __name__ == '__main__':
    longitude, minimo, maximo, media = sequence_status(input_sec=sec)
    print('-'*25)
    print('-longitude:{}'.format(longitude))
    print('-minimo:{}'.format(minimo))
    print('-maximo:{}'.format(maximo))
    print('-media:{}'.format(media))
    print('-'*25)
```
